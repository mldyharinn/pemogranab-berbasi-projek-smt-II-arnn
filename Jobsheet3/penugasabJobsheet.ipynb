```python
class Student:
    def __init__(self, name, score):
        # atribut privat
        self.__name = name
        self.__score = 0
        self.__grade = ""

        # langsung pakai setter supaya divalidasi
        self.score = score

    # ===== PROPERTY NAME =====
    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, value):
        if value.strip() != "":
            self.__name = value
        else:
            print("Nama tidak boleh kosong!")

    # ===== PROPERTY SCORE =====
    @property
    def score(self):
        return self.__score

    @score.setter
    def score(self, value):
        if 0 <= value <= 100:
            self.__score = value
            self.__hitung_grade()  # update otomatis
        else:
            print("Nilai harus di antara 0 sampai 100!")

    # ===== METHOD PRIVATE HITUNG GRADE =====
    def __hitung_grade(self):
        if self.__score >= 90:
            self.__grade = "A"
        elif self.__score >= 80:
            self.__grade = "B"
        elif self.__score >= 70:
            self.__grade = "C"
        elif self.__score >= 60:
            self.__grade = "D"
        else:
            self.__grade = "E"

    # ===== MENAMPILKAN DATA =====
    def show_info(self):
        print("===== DATA MAHASISWA =====")
        print(f"Nama Mahasiswa : {self.__name}")
        print(f"Nilai          : {self.__score}")
        print(f"Grade          : {self.__grade}")

    # ===== DESTRUKTOR =====
    def __del__(self):
        print(f"Data mahasiswa {self.__name} telah dihapus dari sistem.")


# ===== PROGRAM UTAMA =====
if __name__ == "__main__":
    # membuat objek
    data_mhs = Student("Arin", 100)

    # tampilkan data awal
    data_mhs.show_info()

    print("\nNilai diubah ...\n")

    # ubah nilai
    data_mhs.score = 93

    # tampilkan lagi
    data_mhs.show_info()

    # hapus objek
    del data_mhs
```
