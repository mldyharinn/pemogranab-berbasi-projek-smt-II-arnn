```python
from abc import ABC, abstractmethod

# Kelas Abstrak
class Senjata(ABC):
    def __init__(self, nama):
        self.nama = nama

    @abstractmethod
    def serang(self):
        pass

    @property
    @abstractmethod
    def status(self):
        pass

    def info_nama(self):
        return f"Nama senjata: {self.nama}"


# Kelas Pedang
class Pedang(Senjata):
    def __init__(self, nama, panjang_bilah):
        super().__init__(nama)
        self.panjang_bilah = panjang_bilah
        self._daya_tahan = 100  # default

    def serang(self):
        print(f"Pedang {self.nama} menebas dengan bilah sepanjang {self.panjang_bilah} cm!")

    @property
    def status(self):
        return self._daya_tahan

    @status.setter
    def status(self, nilai):
        self._daya_tahan = nilai


# Kelas Panah
class Panah(Senjata):
    def __init__(self, nama, jumlah_anak_panah):
        super().__init__(nama)
        self._ammo = jumlah_anak_panah

    def serang(self):
        if self._ammo > 0:
            self._ammo -= 1
            print(f"Panah {self.nama} melesat! Sisa anak panah: {self._ammo}")
        else:
            print(f"Amunisi Panah {self.nama} habis!")

    @property
    def status(self):
        return self._ammo

    @status.setter
    def status(self, jumlah):
        self._ammo = jumlah


# Program Utama
if __name__ == "__main__":
    pedang1 = Pedang("Excalibur", 90)
    panah1 = Panah("Longbow", 3)

    print(pedang1.info_nama())
    pedang1.serang()
    print("Daya tahan:", pedang1.status)

    print("\n--- Panah ---")
    print(panah1.info_nama())

    panah1.serang()
    panah1.serang()
    panah1.serang()
    panah1.serang()  # habis

    print("Sisa ammo:", panah1.status)

    # Uji kelas abstrak
    try:
        s = Senjata("Tidak Bisa")
    except TypeError as e:
        print("\nError:", e)
```
