```python
# kelas induk
class Mahasiswa:
    def __init__(self, nama, nim, ipk):
        self.nama = nama
        self.nim = nim
        self.ipk = ipk

    def tampilkan_info(self):
        print("Nama :", self.nama)
        print("NIM  :", self.nim)
        print("IPK  :", self.ipk)

    def hitung_predikat(self):
        if self.ipk >= 3.5:
            print("Predikat : Dengan Pujian")
        elif self.ipk >= 3.0:
            print("Predikat : Sangat Memuaskan")
        elif self.ipk >= 2.5:
            print("Predikat : Memuaskan")
        else:
            print("Predikat : Cukup")


# kelas anak 1
class MahasiswaSarjana(Mahasiswa):
    def __init__(self, nama, nim, ipk, semester, sks):
        super().__init__(nama, nim, ipk)
        self.semester = semester
        self.sks = sks

    def tampilkan_info(self):
        super().tampilkan_info()
        print("Semester :", self.semester)
        print("SKS      :", self.sks)


# kelas anak 2
class MahasiswaMagister(Mahasiswa):
    def __init__(self, nama, nim, ipk, judul, pembimbing):
        super().__init__(nama, nim, ipk)
        self.judul = judul
        self.pembimbing = pembimbing

    def tampilkan_info(self):
        super().tampilkan_info()
        print("Judul Tesis :", self.judul)
        print("Pembimbing  :", self.pembimbing)


# program utama
m1 = MahasiswaSarjana("Miladiyah Arinal Haq", "4.33.25.2.13", 3.9, 5, 100)
print("Data Mahasiswa Sarjana")
m1.tampilkan_info()
m1.hitung_predikat()

print("\n----------------------\n")

m2 = MahasiswaMagister("Reigza Nasya", "4.33.25.2.19", 3.0, "Pemograman Berbasis Projek", "Pak Handoko")
print("Data Mahasiswa Magister")
m2.tampilkan_info()
m2.hitung_predikat()
```
