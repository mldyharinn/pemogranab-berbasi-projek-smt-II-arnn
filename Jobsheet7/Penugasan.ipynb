```python
class InventarisBarang:
    def __init__(self):
        # a) Inisialisasi atribut protected sebagai dictionary kosong
        self._items = {}
        print("Sistem Inventaris Barang diinisialisasi.")

    def tambah_barang(self, kode, jumlah):
        # b) Implementasi metode tambah_barang
        print(f"\n[Tambah] Mencoba menambah barang: {kode} sebanyak {jumlah}")
        try:
            # Validasi jumlah harus integer positif
            if not isinstance(jumlah, int) or jumlah <= 0:
                raise ValueError(f"Jumlah ({jumlah}) harus berupa angka bulat positif.")
            
            # Cek jika kode barang sudah ada
            if kode in self._items:
                raise KeyError(f"Error: Kode barang '{kode}' sudah ada.")
            
            # Jika valid, tambahkan ke dictionary
            self._items[kode] = jumlah
            print(f" -> Sukses: Barang '{kode}' ditambahkan dengan stok {jumlah}.")

        except ValueError as e:
            print(f" -> ERROR VALIDASI: {e}")
        except KeyError as e:
            # Menangkap KeyError yang di-raise sendiri
            print(f" -> ERROR KUNCI: {e}")

    def ambil_barang(self, kode, jumlah):
        # c) Implementasi metode ambil_barang
        print(f"\n[Ambil] Mencoba mengambil barang: {kode} sebanyak {jumlah}")
        try:
            # Validasi jumlah
            if not isinstance(jumlah, int) or jumlah <= 0:
                raise ValueError(f"Jumlah pengambilan ({jumlah}) harus angka bulat positif.")
            
            # Cek keberadaan kode (akan memicu KeyError jika tidak ada)
            if kode not in self._items:
                raise KeyError(f"Error: Kode barang '{kode}' tidak ditemukan dalam inventaris.")
            
            # Cek kecukupan stok
            stok_sekarang = self._items[kode]
            if jumlah > stok_sekarang:
                raise ValueError(f"Error: Stok barang '{kode}' tidak mencukupi (tersedia: {stok_sekarang}, diminta: {jumlah}).")
            
            # Proses pengurangan
            self._items[kode] -= jumlah
            print(f" -> Sukses: Mengambil {jumlah} {kode}. Sisa stok: {self._items[kode]}.")

        except ValueError as e:
            print(f" -> ERROR PROSES: {e}")
        except KeyError as e:
            print(f" -> ERROR KODE: {e}")

    def cek_stok(self, kode):
        print(f"\n[Cek] Memeriksa stok untuk kode: {kode}")
        try:
            stok = self._items[kode]
        except KeyError:
            print(f" -> Hasil: Barang '{kode}' tidak ditemukan.")
        else:
            print(f" -> Hasil: Jumlah stok '{kode}' adalah {stok}.")
        finally:
            print(f" -> Info: Pengecekan stok untuk kode '{kode}' telah selesai dilakukan.")

# e) Kode Utama
if __name__ == "__main__":
    inv = InventarisBarang()

    # --- Operasi Tambah Barang ---
    inv.tambah_barang("B001", 10)     
    inv.tambah_barang("B001", 5)     
    inv.tambah_barang("B002", -3)     
    inv.tambah_barang("B002", "lima")

    # --- Operasi Ambil Barang ---
    inv.tambah_barang("B003", 20)     
    inv.ambil_barang("B003", 5)       
    inv.ambil_barang("B003", 25)      
    inv.ambil_barang("B099", 5)       
    inv.ambil_barang("B003", 0)       

    # --- Operasi Cek Stok ---
    inv.cek_stok("B001") 
    inv.cek_stok("B088") 
```
