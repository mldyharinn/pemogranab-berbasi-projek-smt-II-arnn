```python
# 1. Variabel
nama = "Arin"
umur = 18
tinggi = 160.5
is_student = True

print("Nama =", nama)
print("Umur =", umur)
print("Tinggi =", tinggi, "cm")
print("Mahasiswa =", is_student)

# 2. Operasi Aritmatika
a = 10
b = 3

print("\nOPERASI ARITMATIKA")
print("Penjumlahan =", a + b)
print("Pengurangan =", a - b)
print("Perkalian =", a * b)
print("Pembagian =", a / b)
print("Pembagian Bulat =", a // b)
print("Modulus =", a % b)
print("Pangkat =", a ** b)

# 3. Operasi Perbandingan
print("\nOPERASI PERBANDINGAN")
print("a > b =", a > b)
print("a < b =", a < b)
print("a == b =", a == b)
print("a != b =", a != b)
print("a >= b =", a >= b)
print("a <= b =", a <= b)

# 4. Operasi Logika
x = True
y = False

print("\nOPERASI LOGIKA")
print("x and y =", x and y)
print("x or y =", x or y)
print("not x =", not x)

# 5. Percabangan
if a > b and b > 0:
    print("\nKondisi terpenuhi")
else:
    print("\nKondisi tidak terpenuhi")
```
