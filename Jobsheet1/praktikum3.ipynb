```python
# 1. IF
nilai = 85
if nilai > 80:
    print("Lulus dengan nilai tinggi")


# 2. IF-ELSE
umur = 17
if umur >= 18:
    print("Boleh buat SIM")
else:
    print("Belum cukup umur")


# 3. IF-ELIF-ELSE
hari = "Rabu"
if hari == "Senin":
    print("Hari Senin")
elif hari == "Selasa":
    print("Hari Selasa")
elif hari == "Rabu":
    print("Hari Rabu")
else:
    print("Hari lainnya")


# 4. Nested IF
suhu = 35
if suhu > 30:
    print("Panas")
    if suhu > 40:
        print("Sangat panas")
    else:
        print("Masih normal")
else:
    print("Sejuk")


# 5. Logika dalam IF
nilai_teori = 75
nilai_praktik = 80

if nilai_teori >= 70 and nilai_praktik >= 70:
    print("Lulus")
elif nilai_teori < 70 and nilai_praktik < 70:
    print("Tidak lulus")
elif nilai_teori < 70:
    print("Perbaiki teori")
else:
    print("Perbaiki praktik")


# 6. Ternary
angka = -5
status = "Positif" if angka > 0 else "Negatif atau Nol"
print(status)
```
