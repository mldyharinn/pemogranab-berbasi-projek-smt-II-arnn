```python
# 1. FOR
for i in range(5):
    print("Perulangan ke-", i)

# 2. FOR list
buah = ["apel", "mangga", "jeruk"]
for item in buah:
    print(item)

# 3. WHILE
count = 0
while count < 5:
    print("count =", count)
    count += 1

# 4. BREAK
for i in range(10):
    if i == 3:
        break
    print(i)

# 5. CONTINUE
for i in range(5):
    if i == 2:
        continue
    print(i)

# 6. NESTED LOOP
for i in range(3):
    for j in range(2):
        print(i, j)

# 7. ELSE pada loop
for x in range(3):
    print(x)
else:
    print("Loop selesai")

# 8. PASS
for i in range(3):
    if i == 1:
        pass
    print(i)
```
