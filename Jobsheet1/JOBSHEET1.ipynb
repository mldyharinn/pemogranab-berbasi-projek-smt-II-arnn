```python
#PENUGASAN PRAKTIKUM1 
#program menghitung BMI
berat =  float(input("Masukan berat badan anda (kg): "))
tinggi =  float(input("Masukan tinggi badan anda (cm): "))

tinggi = tinggi / 100
bmi = berat / (tinggi ** 2)

if bmi > 18.5 :
    kategori = "Underweight"
elif bmi <= 24.9 :
    kategori = "Normal"
elif bmi <= 29.9 :
    kategori = "Overweight"
else:
    kategori = "Obesitas"

print("Nilai BMI =", round(bmi, 2))
print("Kategori =", kategori)
```

```python
#PENUGASAN PRAKTIKUM1 
#Program 2 Cek bilangan genap, ganjil, dan prima
angka = int(input("Masukan sebuah bilangan :"))

if angka % 2 == 0:
    print("Bilangan Genap")
else:
    print("Bilangan Ganjil")
#Cek bilangan prima
if angka > 1:
    prima = True
    for i in range(2, angka):
        if angka % i == 0:
            prima = False
            break
    if prima:
        print("Bilangan Prima")
    else:
        print("Bukan bilangan prima")
else:
    print("Bukan Bilangan prima")
```

```python
#PRAKTIKUM 1 JOBSHEET 1
# 1. Number (int, float, complex)

# Integer
angka_int = 10
print("angka_int =", angka_int, "->", type(angka_int))

# Float
angka_float = 3.14
print("angka_float =", angka_float, "->", type(angka_float))

# Complex
angka_complex = 2 + 3j
print("angka_complex =", angka_complex, "->", type(angka_complex))


# 2. Boolean
is_active = True
print("is_active =", is_active, "->", type(is_active))


# 3. String
teks = "Hello, Python!"
print("teks =", teks, "->", type(teks))


# 4. List
daftar_angka = [1, 2, 3, 4, 5]
print("daftar_angka =", daftar_angka, "->", type(daftar_angka))


# 5. Tuple
koordinat = (10, 20)
print("koordinat =", koordinat, "->", type(koordinat))


# 6. Dictionary
data_mahasiswa = {
    "nama": "Andi",
    "nim": "A11.2022.12345",
    "jurusan": "Teknik Informatika"
}
print("data_mahasiswa =", data_mahasiswa, "->", type(data_mahasiswa))


# 7. Set
himpunan_angka = {1, 2, 3, 2, 1}
print("himpunan_angka =", himpunan_angka, "->", type(himpunan_angka))


# 8. Konversi tipe data
nilai_str = "100"
print("\nnilai_str =", nilai_str, "->", type(nilai_str))

nilai_int = int(nilai_str)
print("nilai_int =", nilai_int, "->", type(nilai_int))
```

```python
#PRAKTIKUM2 JOBSHEET1
# 1. Variabel
nama = "Arin"
umur = 18
tinggi = 160.5
is_student = True

print("Nama =", nama)
print("Umur =", umur)
print("Tinggi =", tinggi, "cm")
print("Mahasiswa =", is_student)

# 2. Operasi Aritmatika
a = 10
b = 3

print("\nOPERASI ARITMATIKA")
print("Penjumlahan =", a + b)
print("Pengurangan =", a - b)
print("Perkalian =", a * b)
print("Pembagian =", a / b)
print("Pembagian Bulat =", a // b)
print("Modulus =", a % b)
print("Pangkat =", a ** b)

# 3. Operasi Perbandingan
print("\nOPERASI PERBANDINGAN")
print("a > b =", a > b)
print("a < b =", a < b)
print("a == b =", a == b)
print("a != b =", a != b)
print("a >= b =", a >= b)
print("a <= b =", a <= b)

# 4. Operasi Logika
x = True
y = False

print("\nOPERASI LOGIKA")
print("x and y =", x and y)
print("x or y =", x or y)
print("not x =", not x)

# 5. Percabangan
if a > b and b > 0:
    print("\nKondisi terpenuhi")
else:
    print("\nKondisi tidak terpenuhi")
```

```python
#PRAKTIKUM3 JOBSHEET1
# 1. IF
nilai = 85
if nilai > 80:
    print("Lulus dengan nilai tinggi")


# 2. IF-ELSE
umur = 17
if umur >= 18:
    print("Boleh buat SIM")
else:
    print("Belum cukup umur")


# 3. IF-ELIF-ELSE
hari = "Rabu"
if hari == "Senin":
    print("Hari Senin")
elif hari == "Selasa":
    print("Hari Selasa")
elif hari == "Rabu":
    print("Hari Rabu")
else:
    print("Hari lainnya")


# 4. Nested IF
suhu = 35
if suhu > 30:
    print("Panas")
    if suhu > 40:
        print("Sangat panas")
    else:
        print("Masih normal")
else:
    print("Sejuk")


# 5. Logika dalam IF
nilai_teori = 75
nilai_praktik = 80

if nilai_teori >= 70 and nilai_praktik >= 70:
    print("Lulus")
elif nilai_teori < 70 and nilai_praktik < 70:
    print("Tidak lulus")
elif nilai_teori < 70:
    print("Perbaiki teori")
else:
    print("Perbaiki praktik")


# 6. Ternary
angka = -5
status = "Positif" if angka > 0 else "Negatif atau Nol"
print(status)
```

```python
#PRAKTIKUM4 JOBSHEET1
# 1. FOR
for i in range(5):
    print("Perulangan ke-", i)

# 2. FOR list
buah = ["apel", "mangga", "jeruk"]
for item in buah:
    print(item)

# 3. WHILE
count = 0
while count < 5:
    print("count =", count)
    count += 1

# 4. BREAK
for i in range(10):
    if i == 3:
        break
    print(i)

# 5. CONTINUE
for i in range(5):
    if i == 2:
        continue
    print(i)

# 6. NESTED LOOP
for i in range(3):
    for j in range(2):
        print(i, j)

# 7. ELSE pada loop
for x in range(3):
    print(x)
else:
    print("Loop selesai")

# 8. PASS
for i in range(3):
    if i == 1:
        pass
    print(i)
```
